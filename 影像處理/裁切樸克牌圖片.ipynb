# 這個範例是用來教大家怎麼製作自己的樸克牌圖片

```python
from PIL import Image
```

```python
# https://zh.wikipedia.org/wiki/%E6%89%91%E5%85%8B%E7%89%8C
img = Image.open("images/2000px-Svg-cards-2.0.svg.png")
```

```python
img1 = img.crop((0, 0, 154, 224))
img1
```

```python
i = 4
n = 3
img2 = img.crop(((n-1) * 153.75, 223.5*i, n*153.75, 223.5*(i+1)))
img2
```

```python
# 定義樸克牌
numbers = range(1, 14)
colors = ['S', 'H', 'C', 'D']
cards = []

# 產生一副樸克牌
for c in colors:
    for n in numbers:
        cards.append(c + str(n))
```

```python
row = {}
row['C'] = 0
row['D'] = 1
row['H'] = 2
row['S'] = 3
row['J'] = 4

for c in cards:
    i = row[c[0]]
    n = int(c.strip('CDHS'))
    card_img = img.crop(((n-1)*153.75, 223.5*i, n*153.75, 223.5*(i+1)))
    card_img.save("images/{}.png".format(c))
```
