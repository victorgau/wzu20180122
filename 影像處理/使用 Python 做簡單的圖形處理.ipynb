```python
# 載入套件
# 使用 PIL 來處理影像
# 使用 Requests 來抓取網路資料

from PIL import Image
import requests
from io import BytesIO
```

## 介紹一下 PIL 的使用

```python
# 讀出第一章圖
im1 = Image.open("images/Emblem_of_Kaohsiung_City.svg.png")
im1
```

```python
# Rotate 45 degree
im1.rotate(45)
```

```python
# To rotate:
im1.rotate(45, expand=True)
```

```python
# 讀出第二張圖
im2 = Image.open("images/Wenzao_Ursuline_University_of_Languages_logo.svg.png")
im2
```

```python
im1.size
```

```python
size = 125, 150
im1.thumbnail(size, Image.ANTIALIAS)
im1
```

```python
im2.thumbnail(size, Image.ANTIALIAS)
im2
```

```python
im3 = Image.new("RGB",(im1.size[0], im1.size[1] + im2.size[1]),"white")
print(im1.size)
print(im2.size)
print(im3.size)
im3.paste(im1,(0,0,im1.size[0],im1.size[1]), im1)
im3.paste(im2,(0, im1.size[1], im1.size[0], im1.size[1]+im2.size[1]), im2)
im3
# To save:
# im3.save(filename,"JPEG")
```

## 使用 Requests 抓網路上的圖片

```python
# 直接從網路抓取文藻的 logo
# https://zh.wikipedia.org/wiki/File:Wenzao_Ursuline_University_of_Languages_logo.svg
r = requests.get('https://upload.wikimedia.org/wikipedia/zh/thumb/3/33/Wenzao_Ursuline_University_of_Languages_logo.svg/308px-Wenzao_Ursuline_University_of_Languages_logo.svg.png')
i = Image.open(BytesIO(r.content))
i
# To save, use i.save()
```

```python
i.rotate(45)
```

```python
i.rotate(45, expand=True)
```
